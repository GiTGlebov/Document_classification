```python
from sklearn.metrics import jaccard_score
def jac_score(x1):
    jaccard_matrix = np.zeros((x1.shape[0], x1.shape[0]))

# intersection = len(np.intersect1d(vector1, vector2))  # Пересечение уникальных элементов
# union = len(np.union1d(vector1, vector2))

    
    for i in range(x1.shape[0]):
        for j in range(i, x1.shape[0]):
            doc1 = np.array(x1[i].toarray())[0]
            # print(doc1)
            doc2 = np.array(x1[j].toarray())[0]
            # print(doc2)
            jaccard = jaccard_score(doc1, doc2, average='micro')
            jaccard_matrix[i, j] = jaccard
            jaccard_matrix[j, i] = jaccard
            print('jaccard',jaccard)
    return jaccard_matrix

print(jac_score(x1))
```

```python

def Kmeans_iter(euclidean_sent,cosine_sent,jac_sent,real_label=real_label,k=2):
    core = 42
    jac_sc = []
    euc_sc = []
    cos_sc = []
    for i in range(100):
        km_euc = KMeans(n_clusters=k, init='k-means++',random_state=core)
        km_cos = KMeans(n_clusters=k, init='k-means++',random_state=core)
        km_jac = KMeans(n_clusters=k, init='k-means++',random_state=core)
        
        km_euc.fit(euclidean_sent)
        km_cos.fit(cosine_sent)
        km_jac.fit(jac_sent)
        # print(km_euc.labels_)
        jac_sc.append(normalized_mutual_info_score(real_label, km_jac.labels_))
        euc_sc.append(normalized_mutual_info_score(real_label, km_euc.labels_))
        cos_sc.append(normalized_mutual_info_score(real_label, km_cos.labels_))
        core += 3
    return euc_sc,cos_sc,jac_sc

```
